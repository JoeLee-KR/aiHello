```python
def mysq(a):
  return a**2

def mysqlist(a, b):
  result =[]
  for i in b:
    result.append( a(i) )
  return result

print( mysqlist(mysq, [2,3,4]) )
print( mysqlist(lambda x: x**2, [4,3,2]) )
```

```python
from functools import reduce
print( reduce(lambda x, y: x + y, [0, 1, 2, 3, 4]) )
print( reduce(lambda x, y: x + y, [3,2,9,2]) )
print( reduce(lambda x, y: x + y, [4,3,2,1,0]) )
print( reduce(lambda x, y: x + y, [2,9,2,3]) )
print( reduce(lambda x, y: y + x, 'abcde') )
```
