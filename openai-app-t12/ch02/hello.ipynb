```python
from dotenv import load_dotenv
import os
#import dotenv

# .env 파일 로드 (파일 이름을 지정하지 않으면 기본적으로 '.env'를 찾음)
load_dotenv("../../_apikeys.env")  # 또는 load_dotenv("config.env")

# 환경변수에서 API 키 읽기
# api_key = os.getenv("TESTKEY")
api_key = os.getenv("Doogie.2ndKey")

# API 키 사용 예시
if api_key:
    print(f"Key loaded: {api_key[:8]}****...ommitted...")  # 보안을 위해 앞 몇자리만 출력
else:
    print("Key not found.")
```

```python
# tuple
atuple = (1,2,3)
print( f"tuple: {type(atuple)}" )
for aa in atuple:
    print(aa, end=' ')
print( type(atuple[0]))

# list (as array)
alist = [1,2,3]
print( f"list: {type(alist)}" )
alist[0]="11s"
for aa in alist:
    print(aa, end=' ')
print( type(alist[0]), type(alist[1]))

# dictionary
adict = {'a':1, 'b':2, 'c':3}
print( f"dictionary: {type(adict)}" )
adict['a']="11s"
for aa in adict:
    print(aa, end=' ')
print( type(adict['a']), type(adict['b']))
```
