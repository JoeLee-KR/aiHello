```python
# tuple, can not change
atuple = (1,2,3)
print( f"tuple ()...: {type(atuple)}" )
#atuple[0]="11s"
for aa in atuple:
    print(aa, type(aa), end=' ')
print(len(atuple))

```

```python
# list (as array)
alist = [1,2,3]
print( f"list []...: {type(alist)}" )
alist[0]="11s"
for aa in alist:
    print(aa, type(aa), end=' ')
print(len(alist))
```

```python
# dictionary
adict = {'a':1, 'b':2, 'c':3}
print( f"dictionary []...: {type(adict)}" )
for aa in adict:
    print(aa,":", adict[aa], ">", type(adict[aa]), end=' ')
print(len(adict))

adict['b']="new11s"
for aa in adict:
    print(aa,":", adict[aa], ">", type(adict[aa]), end=' ')
print(len(adict))
```

```python
# set
aset = {1,2,3}
print( f"set {{}}...: {type(aset)}" )
aset.add("4th")
for aa in aset:
    print(aa, type(aa), end=' ')
print(len(aset))

aset.pop()
print(aset)

aset.clear()
print(aset)

aset.add('11')
aset.add('22')
print(aset)
for aa in aset:
    print(aa)

sorted(aset)
aset.pop()
print(aset)
```
