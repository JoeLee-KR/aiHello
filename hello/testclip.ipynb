```python
condition = [1,2, 4, 4]

for key in condition:
    if key is 2:
        print(f"condition{key} is NULL")
    else:
        print(f"condition{key} is not NULL")
```

```python
condition = [1,2, None, 3, 4]

for key in condition:
    if key is not None:
        print(key)
    else:
        print("NULL")
```
